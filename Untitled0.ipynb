 **1. What is the difference between a function and a method in Python?**
 - **Function**:
A function is a block of reusable code that performs a specific task. It can be defined using the def keyword and can be called by its name.
Functions are not tied to any object or class. They can exist on their own in the global scope or inside a module.

- **Method**: A method is essentially a function that is associated with an object, typically defined within a class. Methods are functions that are bound to the object (or instance) of the class, meaning they can access and modify the object's attributes.
Methods are called on an object and have at least one parameter by convention, usually self, which refers to the instance of the object.

- **Arguments**:
Arguments are the actual values that we pass to the function when we call it. They correspond to the parameters defined in the function's signature.

 **5. What are iterators in Python and how do they differ from iterables?**

 In Python, iterators and iterables are closely related concepts, but they have distinct roles and characteristics.
- **Iterables**: An iterable is any object in Python that can return an iterator. In other words, an iterable is something we can loop over (iterate over) using a for loop or other iteration methods.
An iterable can be a list, tuple, string, dictionary, set, or any other object that implements the __iter__() method or the __getitem__() method.

- **Iterators:**
An iterator is an object that represents a stream of data. It is used to access the elements of an iterable, one at a time, using the __next__() method. An iterator is created from an iterable by calling the iter() function.



 **6. Explain the concept of generators in Python and how they are defined.**
 - Generators in Python are a special type of iterator that allow us to iterate over a sequence of values lazily (i.e., they generate values one at a time as needed, instead of storing all values in memory at once). This makes generators memory-efficient, especially when working with large datasets.

- A generator is defined using a function (with the def keyword) or an expression (using a generator expression), but instead of using return to return a value, the function uses the yield keyword.

 **8. What is a lambda function in Python and when is it typically used?**
-  "A lambda function" in Python is a small anonymous function that is defined using the lambda keyword. Unlike regular functions defined with def, a lambda function does not require a name. It can take any number of arguments, but it can only contain a single expression. The result of this expression is automatically returned when the lambda function is called.

- They are typically used for short-term functions that are not intended to be reused. Common use cases include passing them as arguments to higher-order functions (map(), filter(), sorted()) or for simple, one-line operations where defining a full function would be unnecessary.


 **9. Explain the purpose and usage of the `map()` function in Python.**

- The map() function in Python is used to apply a given function to all items in an iterable (like a list, tuple, etc.). It returns a map object (which is an iterator) that produces the results of applying the function to each item in the iterable. The map() function is often used when we need to perform a transformation on a sequence of data.
- The map() function is perfect for situations where we need to transform the elements of an iterable (list, tuple, etc.) based on some function.

 **10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**
- The map(), reduce(), and filter() functions are all high-order functions that are commonly used for functional programming in Python. Each of them serves a distinct purpose and has different behavior when applied to iterables.

**Difference between `map()`, `reduce()`, and `filter()`:**

- **map()**: Applies a function to each item of an iterable and returns the transformed items.

- **filter()**: Filters elements from an iterable based on a condition and returns only those that meet the condition.

- **reduce()**: Reduces an iterable to a single value by applying a binary function cumulatively.

 **1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.**


```python
def sum_of_even_numbers(numbers):
    return sum(x for x in numbers if x % 2 == 0)

numbers = [47, 11, 42, 13, 8, 16]
result = sum_of_even_numbers(numbers)
print(result)
```

**3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.**

```python
def square_numbers(numbers):
    return [x ** 2 for x in numbers]

input_list = [1, 2, 3, 4, 5]
squared_list = square_numbers(input_list)
print(squared_list)
```

 **4. Write a Python function that checks if a given number is prime or not from 1 to 200.**


```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

for number in range(1, 201):
    if is_prime(number):
        print(number, "is prime")
```

**5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.**

```python
class FibonacciIterator:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1
        self.count = 0
    def __iter__(self):
        return self
    def __next__(self):
        if self.count < self.n:
            result = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return result
        else:
            raise StopIteration

fibonacci_sequence = FibonacciIterator(10)
for number in fibonacci_sequence:
    print(number)
```

 **6. Write a generator function in Python that yields the powers of 2 up to a given exponent.**


```python
def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

exponent = 5
for power in powers_of_2(exponent):
    print(power)
```

 **7. Implement a generator function that reads a file line by line and yields each line as a string.::**


```python
def read_file_line_by_line(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()
filename = "example.txt"
for line in read_file_line_by_line(filename):
    print(line)
```

 **8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple**.


```python
tuples = [(1, 3), (3, 1), (5, 2), (4, 5)]

sorted_tuples = sorted(tuples, key=lambda x: x[1])

print(sorted_tuples)
```

 **9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.**


```python

celsius_temperatures = [0, 10, 20, 30, 40, 50]

def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

print(fahrenheit_temperatures)
```

 **10. Create a Python program that uses `filter()` to remove all the vowels from a given string.**


```python
def is_not_vowel(char):
    return char.lower() not in 'aeiou'
input_string = "Hello, World!"
filtered_string = ''.join(filter(is_not_vowel, input_string))
print(filtered_string)
```

**11. Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and the quantity. The product should be increased by 10,-  if the value of the order is smaller than 100,00 . Write a Python program using lambda and map.**


```python
# Example list of orders: [Order number, Price per item, Quantity]
orders = [
    [34587, 40.95, 4],  # Order 1
    [98762, 56.80, 5],  # Order 2
    [77226, 32.95, 3],  # Order 3
    [77226, 24.99, 3],  # Order 4
]

# Using lambda and map to create the desired output
result = list(map(lambda order:
                  (order[0],
                   (order[1] * order[2] + 10) if order[1] * order[2] < 100 else order[1] * order[2]),
                  orders))

# Print the result
print(result)
```
